## Generate data or load data

### Trial Data

**Hypersphere data points generator with equi-volume (For testing q-circuit)**

In hyper-dimensional sphere, the volume is proportional to $R^n$, where $R$ is the hypersphere radius and $n$ is the dimension. We generate data points using polar coordinate, assuming the points are isotropic (no tendency on phi_s).

For $n$-dimension sphere, we can transform the polar coordinate to Cartesian coordinate using formulas:
$$
\begin{align}
    x_1 &= r\cos(\phi_1)\\
    x_2 &= r\sin(\phi_1)\cos(\phi_2)\\
    x_3 &= r\sin(\phi_1)\sin(\phi_2)\cos(\phi_3)\\
    &.\\
    &.\\
    x_{n-1} &= r\sin(\phi_1)\sin(\phi_2)...\sin(\phi_{n-1})\cos(\phi_n)\\
    x_{n} &= r\sin(\phi_1)\sin(\phi_2)...\sin(\phi_{n-1})\sin(\phi_n)\\
\end{align}
$$

Where $\phi_{i}\in [0,\pi]$ for $i!=n$ and $\phi_{n}\in [0,2\pi]$. Note when randomly generate $(r, \phi_i, \phi_n)$, be careful of generating procedure. Suppose we want to uniformly generate datas in equi-volume:
```python
r = np.random.rand()*radius                             # Wrong, distribution proportional to R^1
r = (np.random.rand()*radius**dimension)**(1/dimension) # Correct, distribution proportional to R^n
```

````python
def generate_hypersphere_points(number, dimension, radius, normalize=False, constant_r=True):
    '''generate data points in a hyper-sphere'''
    point_array = []
    for n in range(number):
        if constant_r:
            r = radius
        else:
            if normalize: r =  np.random.rand()
            else:         r = (np.random.rand()*radius**dimension)**(1/dimension)
        phi_s = np.random.rand(dimension-1)*np.pi # range in [0, pi]
        phi_s[-1] = phi_s[-1]*2 # last phi is in [0, 2*pi]
        point = []
        for i in range(dimension):
            coordinate = r
            for j in range(i):
                coordinate *= np.sin(phi_s[j]) # exclude the first coordinate
            if i != dimension-1:
                coordinate *= np.cos(phi_s[i]) # exclude the last  coordinate
            point.append(coordinate)
        point_array.append(point)
    return np.array(point_array, requires_grad=False)

def generate_hypersphere_classes(X, radius, class_num, normalize=False):
    '''split datas into class_num equal volume'''
    dimension = X.shape[1]
    Y = []
    for i in range(len(X)):
        '''judge with volume'''
        # if normalize: v = 1/class_num
        # else: v = (radius**dimension)/class_num
        # volume = (np.linalg.norm(X[i]))**dimension
        # Y.append(min(int(volume/v), class_num-1))
        '''judge with sign'''
        S = 1
        for j in range(dimension):
            S *= np.sign(X[i][j])
        if S > 0: Y.append(1)
        else:     Y.append(0)
    Y = np.array(Y, requires_grad=False)
    return Y

def generate_hypersphere_data(train_num, test_num, dimension, radius, class_num, normalize=False, constant_r=True):
    X = generate_hypersphere_points(number=(train_num+test_num), dimension=dimension, radius=radius, normalize=normalize, constant_r=constant_r)
    Y = generate_hypersphere_classes(X=X, radius=radius, class_num=class_num, normalize=normalize)
    x_train, y_train, x_test, y_test = X[:train_num], Y[:train_num], X[train_num:], Y[train_num:]
    return x_train, y_train, x_test, y_test

''' if needed more simple data, consider square data'''
def generate_circle_in_square_data(train_num, test_num, dimension=4, radius=1, length=2):
    X = (np.random.rand(train_num+test_num, dimension)-0.5)*length
    Y = []
    for i in range(train_num+test_num):
        if np.sum(X[i]**2)**0.5 <= radius: Y.append(0)
        else: Y.append(1)
    X = np.array(X, requires_grad=False)
    Y = np.array(Y, requires_grad=False)
    return X[:train_num], Y[:train_num], X[train_num:], Y[train_num:]
````

## Circuits and Functions

### Density Matrix and Q-Circuit

````python
def density_matrix(label):
    if label == 0:
        return np.array([[1,0],[0,0]])
    else:
        return np.array([[0,0],[0,1]])

dev_single = qml.device('default.qubit', wires=1)
@qml.qnode(dev_single)
def single_qcircuit(params, x, y):
    params = params.reshape(2,-1)
    weights = params[0].reshape(-1, len(x))
    biases  = params[1].reshape(-1, len(x))  
    for l in range(len(weights)): # layers, each with 3 parameters
        for i in range(len(x)//3):
            angle = weights[l][3*i:3*(i+1)]*x[3*i:3*(i+1)]+biases[l][3*i:3*(i+1)]
            qml.Rot(*angle, wires=0)
            # qml.Rot(*x[3*i:3*(i+1)], wires=0) # reupload data
            # qml.Rot(*w[3*i:3*(i+1)], wires=0) # weight of certain layer
    dm = density_matrix(y)
    return qml.expval(qml.Hermitian(dm, wires=0))

dev_two = qml.device('default.qubit', wires=2)
@qml.qnode(dev_two)
def two_qcircuit(params, x, y):
    params = params.reshape(2,-1)
    weights = params[0].reshape(-1, 2, len(x))
    biases  = params[1].reshape(-1, 2, len(x)) 
    for l in range(len(weights)): # layers, each with 3 parameters
        for i in range(len(x)//3):
            for q in range(2):
                angle = weights[l][q][3*i:3*(i+1)]*x[3*i:3*(i+1)]+biases[l][q][3*i:3*(i+1)]
                qml.Rot(*(angle), wires=q)
        qml.CZ(wires=[0,1]) # Controlled-Z for each layer
    qml.CZ(wires=[0,1]) # Since no need CZ for last layer, implement one more to get identity (CZ*CZ = I)
    dm = density_matrix(y)
    return qml.expval(qml.Hermitian(dm, wires=0))

dev_four = qml.device('default.qubit', wires=4)
@qml.qnode(dev_four)
def four_qcircuit(params, x, y):
    params = params.reshape(2,-1)
    weights = params[0].reshape(-1, 4, len(x))
    biases  = params[1].reshape(-1, 4, len(x)) 
    pattern = 1
    for l in range(len(weights)): # layers, each with 3 parameters
        for i in range(len(x)//3):
            for q in range(4):
                angle = weights[l][q][3*i:3*(i+1)]*x[3*i:3*(i+1)]+biases[l][q][3*i:3*(i+1)]
                qml.Rot(*(angle), wires=q)
        if l != len(weights)-1:
            if pattern == 1:
                qml.CZ(wires=[0,1])
                qml.CZ(wires=[2,3])
                pattern = -1
            else:
                qml.CZ(wires=[1,2])
                qml.CZ(wires=[0,3])
                pattern = 1
    dm = density_matrix(y)
    return qml.expval(qml.Hermitian(dm, wires=0))
````

### Cost, Prediction, Accuracy

````python
def cost(params_alphas, x_batch, y_batch, circuit, qubits):
    loss = 0
    params = params_alphas[:-2*qubits]
    for i in range(len(x_batch)):
        loss  += (1 - circuit(params, x_batch[i], y_batch[i]))**2
    return loss/len(x_batch)

def weighted_cost(params_alphas, x_batch, y_batch, circuit, qubits):
    loss = 0
    params, alphas = params_alphas[:-2*qubits], params_alphas[-2*qubits:].reshape(qubits,2)
    for i in range(x_batch.shape[0]):
        for q in range(qubits):
            for c in range(2):
                loss += ( alphas[q][c]*circuit(params, x_batch[i], c)-(y_batch[i]==c) )**2
    return loss*0.5/len(x_batch)

def prediction(params, x_batch, circuit):
    y_predict = []
    for i in range(len(x_batch)):
        classification = np.zeros(2)
        for c in range(2):
            classification[c] = circuit(params, x_batch[i], c)
        y_predict.append(np.argmax(classification))
    return np.array(y_predict)

def accuracy(y_true, y_predict):
    return np.sum(y_true == y_predict)/len(y_true)
````

### Train Function

````python
def train(x_train, y_train, x_test, y_test, opt, opt_name="Optimizer", 
          cost_func=cost, predict_func=prediction, qubits=1, circuit=single_qcircuit, 
          layer_number=7, epochs=30, batch_size=20, weighted=True):
    print(f'\nStart Training with {opt_name}... ')
    print(f'Total Epochs: {epochs} | Layers: {layer_number} | Train_num: {len(x_train)} | Test_num: {len(x_test)} | Weighted: {weighted}')
    if x_train.shape[1]%3 !=0:
        x_train = np.c_[x_train, np.zeros((len(x_train), 3-x_train.shape[1]%3))] # pad dimension to multiple of 3
        x_test  = np.c_[x_test,  np.zeros((len(x_test) , 3-x_test.shape[1]%3))]  # pad dimension to multiple of 3

    dimension          = x_train.shape[1]
    weight_bias_number = 2 * layer_number * dimension * qubits
    alpha_number       = 2 * qubits # 2 for classes
    params_alphas      = np.array(np.r_[np.random.rand(weight_bias_number), np.random.rand(alpha_number)+1.], requires_grad=True)

    # initial prediction (not trained yet)
    params     = params_alphas[:-2*qubits]
    train_cost = cost_func(params_alphas, x_train, y_train, circuit, qubits)
    train_acc  = accuracy(y_train, predict_func(params, x_train, circuit))
    test_acc   = accuracy(y_test,  predict_func(params, x_test,  circuit)) 
    print(f"Epoch: {0:2d} | Train Cost: {train_cost:6f} | Train Accuracy: {train_acc:6f} | Test Accuracy: {test_acc:6f}")

    costs, train_accs, test_accs = [train_cost], [train_acc], [test_acc]
    for it in range(epochs):
        '''shuffle -> random batch -> optimization'''
        random_index = np.random.permutation(range(len(x_train)))
        for b in range(len(random_index)//batch_size):
            if b == len(random_index)//batch_size-1: batch_index = random_index[b*batch_size:]
            else:                                    batch_index = random_index[b*batch_size:(b+1)*batch_size]
            params_alphas = opt.step(lambda P: cost_func(P, x_train[batch_index], y_train[batch_index], circuit, qubits), params_alphas)
        
        '''average cost and prediction'''
        params     = params_alphas[:-2*qubits]
        train_cost = cost_func(params_alphas, x_train, y_train, circuit, qubits)
        train_acc  = accuracy(y_train, predict_func(params, x_train, circuit))
        test_acc   = accuracy(y_test,  predict_func(params, x_test,  circuit)) 
        costs.append(train_cost)
        train_accs.append(train_acc)
        test_accs.append(test_acc)
        print(f"Epoch: {it+1:2d} | Train Cost: {train_cost:6f} | Train Accuracy: {train_acc:6f} | Test Accuracy: {test_acc:6f}")
    print('Training Finished ...\n')
    return params_alphas, costs, train_accs, test_accs
````

## Training and Plotting

````python
x_train, y_train, x_test, y_test = load_TJET_4var_data(train_num=300, test_num=200)

stepsize = 0.001
optimizers = {
    "Gradient Descent Optimizer": GradientDescentOptimizer(stepsize=stepsize),
    "Adam Optimizer": AdamOptimizer(stepsize=stepsize, beta1=0.9, beta2=0.99, eps=1e-08),
    "Nesterov Momentum Optimizer": NesterovMomentumOptimizer(stepsize=stepsize, momentum=0.9),
    "Momentum Optimizer": MomentumOptimizer(stepsize=stepsize, momentum=0.9)
    }

opt_name = "Momentum Optimizer"
result_hypersphere = train(x_train, y_train, x_test, y_test, opt=optimizers[opt_name], opt_name=opt_name,
                            cost_func=weighted_cost, predict_func=prediction, qubits=2, circuit=two_qcircuit, 
                            layer_number=2, epochs=50, batch_size=10, weighted=True)
````
